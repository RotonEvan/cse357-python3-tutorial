```python
# Toss a coin 5 times
# 0 -> heads
# 1 -> tails
all_tosses = []
for x in range(5):
    toss = random.choice([0,1])
    all_tosses.append(toss)
print(all_tosses)
```

```python
print(len(all_tosses))
print(sum(all_tosses))
print(sum(all_tosses)/len(all_tosses))
```

```python
# mean of 10^5 tosses
all_tosses_2 = []
for x in range(10**5):
    toss = random.choice([0,1])
    all_tosses_2.append(toss)
print(all_tosses_2)
```

```python
print(len(all_tosses_2))
print(sum(all_tosses_2))
print(sum(all_tosses_2)/len(all_tosses_2))
```

```python
print(len(all_tosses_3))
print(sum(all_tosses_3))
print(sum(all_tosses_3)/len(all_tosses_3))
```

## NumPy examples for Q3

```python
import numpy as np

# Generate an array of 100 random numbers uniformly distributed between 0 and 1
data = np.random.uniform(low=0.0, high=1.0, size=100)

# Display the first 10 elements
print("Random Data Sample:", data[:10])
```

```python
# Sort the data in ascending order
sorted_data = np.sort(data)

# Display the first 10 sorted elements
print("Sorted Data Sample:", sorted_data[:10])
```

```python
# Compute the cumulative sum of the sorted data
cumulative_sum = np.cumsum(sorted_data)

# Display the first 10 cumulative sums
print("Cumulative Sum Sample:", cumulative_sum[:10])
```

```python
# Create an array of indices (1 to n)
n = len(sorted_data)
indices = np.arange(1, n + 1)

# Compute the proportion of data points less than or equal to each value
proportions = indices / n

# Display the first 10 proportions
print("Proportions Sample:", proportions[:10])
```

```python
# Reshape the sorted data into a column vector (n x 1 matrix)
sorted_data_matrix = sorted_data.reshape(-1, 1)

# Display the shape of the matrix
print("Sorted Data Matrix Shape:", sorted_data_matrix.shape)
```

```python
# Compute the mean of the sorted data
mean_value = np.mean(sorted_data)

# Subtract the mean from each element (centering the data)
centered_data = sorted_data - mean_value

# Display the first 10 centered data values
print("Centered Data Sample:", centered_data[:10])
```

```python
# Generate multiple datasets (e.g., 5 datasets of 100 random numbers each)
datasets = [np.random.uniform(0, 1, size=100) for _ in range(5)]

print("Datasets Length:", len(datasets))

print("Datasets:", datasets)

# Stack datasets into a 2D array (5 x 100)
stacked_data = np.vstack(datasets)

# Compute the mean across the datasets (row-wise mean)
mean_across_datasets = np.mean(stacked_data, axis=0)

# Display the first 10 means
print("Mean Across Datasets Sample:", mean_across_datasets[:10])
```

### Interpolation

`np.linspace`: Generates evenly spaced numbers over a specified interval.

`np.interp`: Performs one-dimensional linear interpolation.


```python
# Define a common x-axis grid for interpolation
x_grid = np.linspace(0, 1, 1000)

# Interpolate the proportions onto the x_grid
interpolated_proportions = np.interp(x_grid, sorted_data, proportions)

# Display the first 10 interpolated proportions
print("Interpolated Proportions Sample:", interpolated_proportions[:10])
```

```python
# Compute the standard error for the proportions
standard_error = np.sqrt(proportions * (1 - proportions) / n)

# Set the z-score for a 95% confidence interval
z_score = 1.96  # Approximate value for 95% CI

# Calculate the confidence intervals
lower_bound = proportions - z_score * standard_error
upper_bound = proportions + z_score * standard_error

# Ensure the bounds are within [0, 1]
lower_bound = np.maximum(lower_bound, 0)
upper_bound = np.minimum(upper_bound, 1)

# Display the first 10 confidence intervals
print("Confidence Interval Sample:")
for i in range(10):
    print(f"Proportion: {proportions[i]:.3f}, CI: [{lower_bound[i]:.3f}, {upper_bound[i]:.3f}]")
```

```python
import matplotlib.pyplot as plt

# Plot the empirical cumulative distribution function (eCDF)
plt.figure(figsize=(8, 6))
plt.step(sorted_data, proportions, where='post', label='eCDF')

# Plot the confidence intervals
plt.fill_between(sorted_data, lower_bound, upper_bound, step='post', color='lightgray', alpha=0.5, label='95% Confidence Interval')

# Plot the original data points as crosses on the x-axis
plt.plot(sorted_data, np.zeros_like(sorted_data), 'x', label='Data Points')

# Customize the plot
plt.xlabel('Data Values')
plt.ylabel('Proportion')
plt.title('Empirical Cumulative Distribution Function with Confidence Intervals')
plt.legend()
plt.grid(True)
plt.show()
```
